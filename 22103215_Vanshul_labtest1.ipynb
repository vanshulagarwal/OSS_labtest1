```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,4*np.pi,4*np.pi/1440)
y = (1 + np.sin(x))*100
plt.plot(x,y)
```

```python
def introduce_noise(v):
    noise = np.random.rand(len(v))
    return v+(noise*10)

afterNoise = introduce_noise(y)
plt.figure(figsize=(10,10))
plt.plot(x,afterNoise)
```

```python
def calc_hourly_avg(x, v):
    
    x_update=np.split(x, 24)
    vehiclePerHour=np.split(v, 24)
    # print(vehiclePerHour)
    xout = [i[23] for i in x_update]
    avg = np.mean(vehiclePerHour, axis=1)
    return xout,avg
```

```python
x_hourly,hourlyavg = calc_hourly_avg(x,smootheddata)
x_hourly, hourlyavg
```
